```python
x = pyber['fare']
y = pyber['rides']
z = pyber['driver_count']
 
c = ["green", "blue", "yellow"]
plt.scatter(x,y,s=z*4, c=c, alpha=0.4, linewidth=1)


plt.xlabel("Number of Rides (Per City)")
plt.ylabel("Average Fare ($)")
plt.title("Pyber", loc="left")
plt.style.use('seaborn-pastel')

textstr = 'Note: Circle size correlates with number of rides per city.'
# print textstr
plt.text(55, 35, textstr, fontsize=10)
plt.subplots_adjust(left=0.25)
plt.grid(True)


plt.show()
```
